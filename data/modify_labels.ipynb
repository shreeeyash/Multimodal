```python
import h5py
import numpy as np
import pandas as pd
```

```python
ey_train = h5py.File("ey_train.h5","r")
print(list(ey_train.keys()))
ey_train = np.array(ey_train.get('d1'))
ey_train.shape
df = pd.DataFrame(ey_train,columns=["anger", "disgust", "fear", "happy", "sad", "surprise"])
print("\nBEFORE PREPROCESSING")
print(df.astype(bool).sum(axis=0))
print((df>3).sum())
df.loc[df["happy"]>3,"happy"] = 3
print("\nAFTER VALUE CLIPPING")
df.describe()
```

```python
for i in range(df.shape[0]):
    col = df.iloc[i].idxmax()
    df.loc[i,col]=1
    df.loc[i,df.columns != col] = 0
```

```python
df
```

```python
df.to_csv('y_train_onehot.csv')
```
